```python
%sql
with cleaned_income as (
select *, 
       cast(regexp_replace(ANNUALINCOME, '[\\$,]', '') as int) as ANNUALINCOME_INT,
       cast(regexp_replace(ANNUALINCOME, '[$,]' , '') as int ) as Final_income   
from snow_pro_dev.snowpro_l0.adventureworks_customers)

select max(final_income) from cleaned_income
where final_income < ( select max(final_income) from cleaned_income )
```

```python
%sql
with change1 
( select * , cast (regexp_replace (ANNUALINCOME, '[$,]' , '') as int) as income
 from snow_pro_dev.snowpro_l0.adventureworks_customers)

 select max(income) from change1
 where income < (select max(income) from change1)
 
```

```python
%sql
select product_id, profit, 
row_number() over (partition by product_id order by profit desc) as row_rnk,
rank() over(partition by product_id order by profit desc) as rnk,
dense_rank() over(partition by product_id order by profit desc) as dnse_rnk
from esat_snow_dev.sales_l0.sales

-- order by product_id, profit desc
```

```python
%sql
select rnk.*,prd.product_name,category from

(select product_id, customer_id, profit,

row_number() over(partition by product_id order by profit desc) as dnse_rnk

from esat_snow_dev.sales_l0.sales) rnk

left join esat_snow_dev.sales_l0.products prd on rnk.product_id = prd.product_id

where rnk.dnse_rnk in (1,2)
```
